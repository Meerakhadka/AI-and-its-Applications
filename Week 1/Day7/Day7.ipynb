#Basic plotting with matplotlib


```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [5,6,7,8,9]
plt.plot(x,y, marker="o")
```

```python
#creating a parabola
import matplotlib.pyplot as plt
x = [-4,-3,-2,-1,0,1,2,3,4]
y = [16,9,4,1,0,1,4,9,16]
plt.plot(x,y, marker="o")
```

```python
#
import matplotlib.pyplot as plt
x = [0,1,2,3,4]
y = [0,1,4,9,16]
plt.plot(x,y, marker="o")
```

```python
#Creating a line plot and adding labels and title
x = [0,1,2,3,4]
y = [0,1,4,9,16]

plt.title("Data")
plt.xlabel("x data")
plt.ylabel("y data")

plt.plot(x, y)
plt.savefig("plt.png")
plt.grid(True)
plt.show()
```

#Exercise: visualizing a dataframe with matplotlib

```python
#
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Score": [25, 30, 35, 40],
}
df = pd.DataFrame(data)
plt.bar(df["Name"], df["Score"])
```

```python
# Task 6: Create a DataFrame with columns ”City” ([”London”, ”Paris”,
# ”Tokyo”]) and ”Population” ([9.0, 2.1, 37.4]). Create a bar plot
# of Population vs. City, add labels and a title, and save it as populationplot.png.

import pandas as pd
import matplotlib.pyplot as plt
# Create the DataFrame
data = {
    "city": ["london", "paris", "tokyo"],
    "Population": [9.0, 2.1, 37.4],
}
# Create the bar plot
plt.title("Population of cities")
plt.xlabel("city")
plt.ylabel("population")
df = pd.DataFrame(data)
plt.bar(df["city"], df["Population"])
plt.savefig("populationplot.png")
plt.show()
```

```python
# Task 5: Create a line plot with x-values [0, 1, 2, 3] and y-values [0, 1, 4, 9]. Add
# a title, labels, and grid, then save it as squareplot.png.

import matplotlib.pyplot as plt
x = [0, 1, 2, 3]
y = [0, 1, 4, 9]
plt.title("square plot")
plt.xlabel("x data")
plt.ylabel("y data")
# Create the line plot
plt.plot(x, y, marker='o')
plt.grid(True)
plt.savefig("squareplot.png")
plt.show()

```
